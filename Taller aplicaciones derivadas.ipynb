## Problema del pez nadando río arriba
---

### Paso 1: Dibujo y asignación de variables

- Velocidad del pez respecto al agua: \( v \)
- Velocidad de la corriente: \( v_c \)
- Velocidad efectiva del pez respecto al suelo (contra corriente):

$$
v_{contracorriente} = v - v_c
$$

- Distancia recorrida: \( d \)
- Tiempo en recorrer \( d \):

$$
t = \frac{d}{v - v_c}
$$

---

### Paso 2: Función objetivo \( E(v) \)

La energía es proporcional al producto del tiempo y el cubo de la velocidad del pez:

$$E(v) = t \cdot v^3$$

Sustituimos el tiempo:

$$E(v) = k \cdot \frac{d}{v - v_c} \cdot v^3 = kd \cdot \frac{v^3}{v - v_c}$$

---

### Paso 3: Minimizar

Minimizamos la función:

$$f(v) = \frac{v^3}{v - v_c}$$

Derivamos usando la regla del cociente:

$$f'(v) = \frac{(3v^2)(v - v_c) - v^3(1)}{(v - v_c)^2}$$

Simplificamos el numerador:

$$f'(v) = \frac{3v^3 - 3v^2 v_c - v^3}{(v - v_c)^2} = \frac{2v^3 - 3v^2 v_c}{(v - v_c)^2}$$

Buscamos puntos críticos:

$$
2v^3 - 3v^2 v_c = 0 \\
v^2(2v - 3v_c) = 0
$$

Soluciones:

- $$ v = 0 $$ (no válida)
- $$ v = \frac{3}{2}v_c $$

---

### Conclusión:

La velocidad v que minimiza la energía gastada por el pez al nadar contra la corriente es:

$$
v = \frac{3}{2}v_c
$$



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**5 - 6*(x**3) + 2*x - 2

def fprima(x):
    return 5*x**4 - 18*x**2 + 2

tol = 0.0001
n = 1000
x = 1


def metodoNewton(x, tol, n):

    for i in range(n-1):
        x1 = x - f(x) / fprima(x)

        if x1 - x < tol:
            print("x", i+1, "x1:", x1, "Fin")
            break
        x = x1
        print("x", i+1, "x1", x1)
    return x1


pasos = metodoNewton(x, tol, n)

x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x)")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```
